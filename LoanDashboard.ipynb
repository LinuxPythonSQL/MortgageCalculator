```python
from tkinter import *
```

```python
class LoanDashboard: 
    
    def __init__(self): 
        window = Tk()
        window.title("Loan Officer Dashboard")
        #Creating the labels for the inputs and outputs
        Label(window, text ="Appraised Value").grid(row = 1, column = 1, sticky = W)
        Label(window, text ="Loan Amount").grid(row = 2, column = 1, sticky = W)
        Label(window, text ="Interst Rate").grid(row = 3, column = 1, sticky = W)
        Label(window, text ="Years").grid(row = 4, column = 1, sticky = W)
        Label(window, text ="Principal + Interest").grid(row = 5, column = 1, sticky = W)
        Label(window, text ="Home Insurance").grid(row = 6, column = 1, sticky = W)
        Label(window, text ="Property Tax").grid(row = 7, column = 1, sticky = W)
        Label(window, text ="HOA").grid(row = 8, column = 1, sticky = W)
        Label(window, text ="Other").grid(row = 9, column = 1, sticky = W)
        Label(window, text ="Total Payment").grid(row = 10, column = 1, sticky = W)

        Label(window, text = "Gross Monthly Income").grid(row = 1, column = 3, sticky = W)
        Label(window, text = "Other Income").grid(row = 2, column = 3, sticky = W)
        Label(window, text = "Total Income").grid(row = 3, column = 3, sticky = W)
        Label(window, text = "Credit Card Payments").grid(row = 4, column = 3, sticky = W)
        Label(window, text = "Car Payment").grid(row = 5, column = 3, sticky = W)
        Label(window, text = "Other Debt").grid(row = 6, column = 3, sticky = W)
        Label(window, text = "Total Debt").grid(row = 7, column = 3, sticky = W)
        
        Label(window, text = "Credit Score").grid(row = 1, column = 5, sticky = W)
        Label(window, text = "Loan to Value Ratio").grid(row = 2, column = 5, sticky = W)
        Label(window, text = "Front-end DTI Ratio").grid(row = 3, column = 5, sticky = W)
        Label(window, text = "Back-end DTI Ratio").grid(row = 4, column = 5, sticky = W)

        #creating inputs and output for just the mortgage
        self.appraisedValueVar = StringVar() 
        Entry(window, textvariable = self.appraisedValueVar, justify = RIGHT).grid(row = 1, column = 2)
        self.loanAmountVar = StringVar() 
        Entry(window, textvariable = self.loanAmountVar, justify = RIGHT).grid(row = 2, column = 2)
        self.interestRateVar = StringVar() 
        Entry(window, textvariable = self.interestRateVar, justify = RIGHT).grid(row = 3, column = 2)
        self.yearsVar = StringVar()
        Entry(window, textvariable = self.yearsVar, justify = RIGHT).grid(row = 4, column = 2)
        self.principalInterestVar = StringVar() 
        #What the Principal + Interest Paymet should be 
        lblPrincipalInterest = Label(window, textvariable = self.principalInterestVar,
                                    font=("Arial", 11)).grid(row = 5, column = 2, sticky = E)
        
        #inputs to add more to the housing expense to come up with the monthly total (PITI)
        self.homeInsuranceVar = StringVar()
        Entry(window, textvariable = self.homeInsuranceVar, justify = RIGHT).grid(row = 6, column = 2)
        self.propertyTaxVar = StringVar()
        Entry(window, textvariable = self.propertyTaxVar, justify = RIGHT).grid(row = 7, column = 2)
        self.hoaVar = StringVar()
        Entry(window, textvariable = self.hoaVar, justify = RIGHT).grid(row = 8, column = 2)
        self.otherVar = StringVar()
        Entry(window, textvariable = self.otherVar, justify = RIGHT).grid(row = 9, column = 2)
        self.totalPaymentVar = StringVar() 
        lblTotalPayment = Label(window, 
                                textvariable = self.totalPaymentVar, 
                                font=("Arial", 11)).grid(row = 10, column = 2, sticky = E) 
        
        #Gross monthly income = income before taxes
        self.grossMonthlyIncomeVar = StringVar() 
        Entry(window, textvariable = self.grossMonthlyIncomeVar, justify = RIGHT).grid(row = 1, column = 4)
        #other income like rent, annuity, dividends or other sources of income
        self.otherIncomeVar = StringVar() 
        Entry(window, textvariable = self.otherIncomeVar, justify = RIGHT).grid(row = 2, column = 4)
        self.totalIncomeVar = StringVar() 
        lblTotalIncome = Label(window, 
                               textvariable = self.totalIncomeVar,
                               font=("Arial", 11)).grid(row = 3, column = 4, sticky = E)
        
        self.creditCardPaymentsVar = StringVar() 
        Entry(window, textvariable = self.creditCardPaymentsVar, justify = RIGHT).grid(row = 4, column = 4)
        self.carPaymentVar = StringVar() 
        Entry(window, textvariable = self.carPaymentVar, justify = RIGHT).grid(row = 5, column = 4)
        self.otherDebtVar = StringVar() 
        Entry(window, textvariable = self.otherDebtVar, justify = RIGHT).grid(row = 6, column = 4)
        self.totalDebtVar = StringVar() 
        lblTotalDebt = Label(window, 
                             textvariable = self.totalDebtVar,
                             font=("Arial", 11)).grid(row = 7, column = 4, sticky = E)
        
        #this will be used later in the future
        self.creditScoreVar = StringVar()
        Entry(window, textvariable = self.creditScoreVar, justify = RIGHT).grid(row = 1, column = 6)
        #Loan to value (LTV) = Loan Amount / Appraised Value
        self.ltvVar = StringVar() 
        lblLTV = Label(window, 
                       textvariable = self.ltvVar,
                       font=("Arial", 11)).grid(row = 2, column = 6, sticky = E)
        #Front End Debt-to-income (DTI) Ratio = Total Payment / Total Income
        self.fDtiVar = StringVar() 
        lblFDti = Label(window, 
                        textvariable = self.fDtiVar,
                        font=("Arial", 11)).grid(row = 3, column = 6, sticky = E)
        #Back End DTI Ratio = (Tptal Payment + Total Debt) / Total Income
        self.bDtiVar = StringVar() 
        lblBDti = Label(window, 
                        textvariable = self.bDtiVar,
                        font=("Arial", 11)).grid(row = 4, column = 6, sticky = E)
        
        #create the button 
        btComputeEverything = Button(window, 
                                     text = "Compute Everything",
                                     command = self.computeEverything).grid(row = 10, column = 6, sticky = E)  
        
        
        window.mainloop()
 # compute everything.
    def computeEverything(self):
        #This will calculate the monthly Principal and Interest Payment
        monthlyPayment = self.getMonthlyPayment( 
        float(self.loanAmountVar.get()), 
        float(self.interestRateVar.get()) / 1200, 
        int(self.yearsVar.get())) 
        
        self.principalInterestVar.set(format(monthlyPayment, '10.2f')) 
        
        #PITI
        totalPayment = float(self.principalInterestVar.get()) + float(self.homeInsuranceVar.get()) + \
        float(self.propertyTaxVar.get()) + float(self.hoaVar.get()) + float(self.otherVar.get())                 
        self.totalPaymentVar.set(format(totalPayment, '10.2f'))
        
        totalIncome = float(self.grossMonthlyIncomeVar.get()) + float(self.otherIncomeVar.get())
        self.totalIncomeVar.set(format(totalIncome, '10.2f'))
        
        totalDebt = float(self.creditCardPaymentsVar.get()) + float(self.carPaymentVar.get()) +\
        float(self.otherDebtVar.get())
        self.totalDebtVar.set(format(totalDebt, '10.2f'))
        
        #Loan-to-Value
        ltv = float(self.loanAmountVar.get()) / float(self.appraisedValueVar.get())
        self.ltvVar.set(format(ltv, '10.2%'))
        
        #Front-end DTI
        fDti = float(self.totalPaymentVar.get()) / float(self.totalIncomeVar.get())
        self.fDtiVar.set(format(fDti, '10.2%'))
        
        #Back-end DTI
        bDti = (float(self.totalPaymentVar.get()) + float(self.totalDebtVar.get())) / \
                float(self.totalIncomeVar.get())
        self.bDtiVar.set(format(bDti, '10.2%'))
    
    def getMonthlyPayment(self, loanAmount, monthlyInterestRate, numberOfYears):  
        # compute the monthly payment. 
        monthlyPayment = loanAmount * monthlyInterestRate / (1
        - 1 / (1 + monthlyInterestRate) ** (numberOfYears * 12)) 
        return monthlyPayment; 
        root = Tk()
    
# call the class to run the program. 
LoanDashboard() 
```
